```python
%matplotlib inline
import pandas as pd
import numpy as np


import matplotlib.pyplot as plt

from sklearn import feature_selection
import matplotlib.pyplot as plt


""" SEE http://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.f_regression.html

Univariate linear regression tests

Quick linear model for testing the effect of a single regressor, 
sequentially for many regressors.

This is done in 3 steps: 
1. the regressor of interest and the data are orthogonalized wrt constant regressors 
2. the cross correlation between data and regressors is computed 
3. it is converted to an F score then to a p-value"""

from sklearn.feature_selection.univariate_selection import f_regression
```

```python
cars = pd.read_csv("cars_complex.csv")
print cars.head()
```

```python

# create an array that includes features that might be relevant; exclude the response MPG.city
features = cars[['Horsepower','EngineSize','RPM','Rev.per.mile']]

# 
f, p = f_regression(features, cars['MPG.city'].values)

print "\n F score:\n",f

print " p-value:\n",p
```

```python
# what is max value in regressed F score?
np.amax(f)
```

```python
# what is the index of the max value F score
f.argmax()
```

```python
# what is name of the feature that has that max value F score
features.columns[f.argmax()]
```
