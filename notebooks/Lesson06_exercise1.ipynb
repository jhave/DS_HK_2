```python
%matplotlib inline
import pandas as pd
import numpy as np

from sklearn import feature_selection
from sklearn import linear_model

import matplotlib.pyplot as plt


cars = pd.read_csv("cars.csv")
#print cars.head()

plt.scatter(cars["speed"],cars["dist"])
plt.show()
```

```python
regr  = linear_model.LinearRegression()
```

```python
speed = [[x] for x in cars['speed'].values]
dist = cars['dist'].values
```

```python
regr.fit(speed,dist)
```

```python
print regr.coef_
```

```python
print np.mean((regr.predict(speed)-dist)**2)
```

```python
print regr.score(speed,dist)
```

```python
#plt.figure()
plt.scatter(speed,dist)
plt.plot(speed,regr.predict(speed))
```
